# Investigaci√≥n de Operaciones

## Problema a estudiar:
problema.png

### Datos:

| Caramelo (kg) | Precio venta (Euro) | Az√∫car (gr) | Fruta (gr) |
|:-------------:|:-------------------:|:-----------:|:----------:|
|       $x_1$      |         200         |     100     |     200    |
|       $x_2$      |         300         |     400     |     400    |

### Restricciones

* La diferencia entre la producci√≥n semanal de C1 y C2 no sea inferior a 5 kg.
  * **Restricci√≥n 1**: $ x_1 - x_2 \geq 5$

* Adem√°s, el contenido de fruta semanal debe ser al menos de 1600 g.
  * **Restricci√≥n 2**: $ 200x_1 + 400x_2 \geq 1600 $

* Problema de programaci√≥n lineal.
  * $ x_1 \geq 0 $
  * $ x_2 \geq 0 $

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,10,100)
```

```python
def grafico():
  plt.ylim([0,5])
  plt.xlim([0,10])
  plt.xlabel("x1")
  plt.ylabel("x2")
  plt.grid()
```

### Restricci√≥n 1
La diferencia entre la producci√≥n semanal de C1 y C2 no sea inferior a 5 kg.

$ x_1 - x_2 \geq 5 $


```python
grafico()
plt.title("Restricci√≥n 1")
plt.plot(x, x-5)
plt.fill_between(x,x-5,0,color="azure")
```

### Restricci√≥n 2
El contenido de fruta semanal debe ser al menos de 1600 g.

$ 200x_1 + 400x_2 \geq 1600 $

1. Divido por 400: $ \frac{x_1}{2} + x_2 \geq 4 $
2. Con $ x_2 = 0 $ queda como: $ \frac{x_1}{2} \geq 4 $
3. Invirtiendo s√≠mbolo para graficar mejor: $ -\frac{x_1}{2} \leq 4 $

```python
grafico()
plt.title("Restricci√≥n 2")
plt.plot(x, -x/2+4, color="orange")
plt.fill_between(x, -x/2+4, 100, color="azure")
```

### Ambas restricciones

```python
grafico()
plt.title("Restricci√≥n 1 y 2")
plt.plot(x, x-5)
plt.plot(x, -x/2+4)
plt.fill_between(x, x-5, -x/2+4, x>6, color="azure")
```

## Pregunta A
Buscar las soluciones eficientes que **maximicen el ingreso** y
**minimicen el contenido de fruta** de la producci√≥n semanal.

### Modelo

$ \text{Max} (\text{euros},-\text{fruta}) = \text{Max} (200x_1 + 300x_2, -100x_1 - 400x_2 ) $

#### Punto (8,0)
$ 	\begin{array}{ll}
		x_1 = 8 \\
		x_2 = 0
	\end{array}
  \Rightarrow (200 \times 8, -100\times8) = (1600,-800)
 $

#### Punto (6,1)
$ 	\begin{array}{ll}
		x_1 = 6 \\
		x_2 = 1
	\end{array}
  \Rightarrow (200 \times 6 + 300 \times 1, -100 \times 6 - 400\times 1) = (1500,-1000)
 $
